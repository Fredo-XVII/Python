```python
# Select columns
hr_year = batting_2000[keep_columns_hr]
hr_year.head()
hr_year.tail()
```

```python
# Plot data
plt_data = hr_year.groupby('yearID').sum()
plt_data.reset_index(inplace=True)
plt_data.head()
```

```python
# plt.axis([xmin,xmax, ymin, ymax])
plt.axis([2000,2016, 4000, 6000])
plt.title('MLB Home Runs by Year')

x_data = list(plt_data['yearID'])
y_data = list(plt_data['HR'])

plt.title('MLB ANNUAL NUMBER OF HOME RUNS')
plt.xlabel('YEAR')
plt.ylabel('NUMBER OF HOME RUNS')
# colors r g b c m y k w
plt.plot(x_data, y_data, color = 'b', linestyle = '', marker = 'o')

plt.show()
```

```python
# ADD LEAGUE TO AGGREGATION
plt_data_lg = hr_year.groupby(['lgID','yearID']).sum()
plt_data_lg.reset_index(inplace = True)
plt_data_lg.head()
```

```python
plt_data_al = plt_data_lg[plt_data_lg.lgID == 'AL']
plt_data_nl = plt_data_lg[plt_data_lg.lgID == 'NL'] 
```

```python
%pylab inline
pylab.rcParams['figure.figsize'] = (12, 8)

plt.plot(plt_data_nl.yearID, plt_data_nl.HR, color = 'b', linestyle = '', marker = '+')
plt.plot(plt_data_al.yearID, plt_data_al.HR, color = 'r', linestyle = '', marker = 'x')
# markerfacecolor = 'green', markeredgecolor= 'red') # part of the plot function...more options
plt.show()
```
